```python
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import pandas as pd
def safelog2(x):
    if x == 0:
        return(0)
    else:
        return(np.log2(x))
```

### Importing Data

```python
# read data into memory
data_set = np.genfromtxt("hw05_data_set_train.csv", delimiter = ",")
# get X and y values for training data
x_train = data_set[:, 0]
y_train = data_set[:, 1]

# read data into memory
data_set = np.genfromtxt("hw05_data_set_test.csv", delimiter = ",")
# get X and y values for test data
x_test = data_set[:,0]
y_test = data_set[:,1]

# get number of classes, number of samples, and number of features
K = np.max(y_train)
N = x_train.shape[0]
#D = y_train.shape[1]

# get numbers of train and test samples
N_train = len(y_train)
N_test = len(y_test)
```

### Decision Tree Regression Algorithm

```python
def decision_tree_reg_alg(P):
    # Initilization
    # create necessary data structures
    node_indices = {}
    is_terminal = {}
    need_split = {}
    
    node_means = {}
    node_splits = {}

    # put all training instances into the root node
    node_indices[1] = np.array(range(N_train))
    is_terminal[1] = False
    need_split[1] = True
    
    # Tree inference
    while True:
        # find nodes that need splitting
        split_nodes = [key for key, value in need_split.items()
                       if value == True]
        # check whether we reach all terminal nodes
        if len(split_nodes) == 0:
            break
        # find best split positions for all nodes
        for split_node in split_nodes:
            data_indices = node_indices[split_node]
            need_split[split_node] = False
            node_mean = np.mean(y_train[data_indices])
            # Pruning condition
            if x_train[data_indices].size <= P:
                is_terminal[split_node] = True
                node_means[split_node] = node_mean ## update the mean
            else:
                is_terminal[split_node] = False
                
                unique_values = np.sort(np.unique(x_train[data_indices]))
                split_positions = (unique_values[1:len(unique_values)] + unique_values[0:(len(unique_values) - 1)]) / 2
                split_scores = np.repeat(0.0, len(split_positions))
                for s in range(len(split_positions)):
                    left_indices = data_indices[x_train[data_indices] < split_positions[s]]
                    right_indices = data_indices[x_train[data_indices] >= split_positions[s]]
                    
                    total_err = 0
                    if len(left_indices) > 0:
                        total_err += np.sum((y_train[left_indices] - np.mean(y_train[left_indices])) ** 2)
                    if len(right_indices) > 0:
                        total_err += np.sum((y_train[right_indices] - np.mean(y_train[right_indices])) ** 2)
                    split_scores[s] = total_err / (len(left_indices)+len(right_indices))
                    
                if len(unique_values) == 1 :
                    is_terminal[split_node] = True
                    node_means[split_node] = node_mean
                    continue
                
                best_split = split_positions[np.argmin(split_scores)]
                node_splits[split_node] = best_split
                
                # create left node using the selected split
                left_indices = data_indices[(x_train[data_indices] < best_split)]
                node_indices[2 * split_node] = left_indices
                is_terminal[2 * split_node] = False
                need_split[2 * split_node] = True

                # create right node using the selected split
                right_indices = data_indices[(x_train[data_indices] >= best_split)]
                node_indices[2 * split_node + 1] = right_indices
                is_terminal[2 * split_node + 1] = False
                need_split[2 * split_node + 1] = True
    return node_splits,node_means,is_terminal
```

###  Performance Algoithm

```python
def performance(X, node_splits, node_means, is_terminal):
    # traverse tree for training data points
    y_predicted = []
    for i in range(len(X)):
        index = 1 #start from root
        while True:
            if is_terminal[index] == True:
                y_predicted.append(node_means[index])
                break
            if X[i] > node_splits[index]:
                ## go to right child
                index = index * 2 + 1
            else:
                ## go to left child
                index = index * 2
    return y_predicted
```

### Parameters

```python
minimum_value = 0.0
maximum_value = 2.0
data_interval = np.linspace(minimum_value, maximum_value, 400)
P = 30
```

### Plot

```python
## For the training set
fig, ax = plt.subplots(2, 1, figsize = (13, 8)) 
plt.subplot(2, 1, 1)
node_splits, node_means, is_terminal = decision_tree_reg_alg(P)
pred_train = np.array(performance(x_train, node_splits, node_means, is_terminal))

my_plot = np.array(performance(data_interval, node_splits, node_means, is_terminal))

## Plot
plt.plot(x_train,y_train, "b.", markersize = 10, label = "training")
plt.plot(data_interval, my_plot, color = "black")
plt.xlabel("Time (sec)")
plt.ylabel("Signal (millivolt)")  
plt.legend(prop={'size': 12})

## For the test set
plt.subplot(2, 1, 2)
node_splits, node_means, is_terminal = decision_tree_reg_alg( P)
pred_test = np.array(performance(x_test, node_splits, node_means, is_terminal))
int_plot = performance(data_interval, node_splits, node_means, is_terminal)

## Plot
plt.plot(x_test, y_test, "r.", markersize = 10, label = "test")
plt.plot(data_interval, int_plot, color = "black")
plt.xlabel("Time (sec)")
plt.ylabel("Signal (millivolt)")  
plt.legend(prop = {'size': 12})

plt.show()
```

### RMSE Algorithm

```python
def rmse(y, pred):
    return np.sqrt(sum((y - pred) ** 2) / len(y))

#calculate and print RMSE
print("RMSE on training set is", rmse(y_train, pred_train), "when P is", P)
print("RMSE on test set is", rmse(y_test, pred_test), "when P is", P)
```

### Decision Tree on Different P values

```python
rmse_values = []
my_range = range(10, 55, 5)
for P in my_range:
    node_splits, node_means, is_terminal = decision_tree_reg_alg(P)
    pred = np.array(performance(x_train, node_splits, node_means, is_terminal))
    rmse_values.append(rmse(y_train, pred))
rmses = np.array(rmse_values)
fig = plt.figure(figsize = (13, 5))
plt.plot(my_range, rmse_values, "b.-", markersize = 15, label = "training")

rmse_values = []
for P in my_range:
    node_splits, node_means, is_terminal = decision_tree_reg_alg(P)
    pred = np.array(performance(x_test, node_splits, node_means, is_terminal))
    rmse_values.append(rmse(y_test, pred))
rmses = np.array(rmse_values)
plt.plot(my_range, rmse_values, "r.-", markersize = 15, label = "test")

plt.legend(prop = {'size': 12})
plt.xlabel("Pre-pruning size (P)")
plt.ylabel("RMSE")
plt.show()
```
